```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings= (
    OllamaEmbeddings(model="gemma:2b")
)
```

```python
r1=embeddings.embed_documents(
    [
        "Alpha is the first letter of Greek word",
        "Beta is the second letter of Greek alphabet"
    ]
)
```

```python
print(r1[0])
print(r1[1])
```

```python
embeddings.embed_query("What is the second lettter of Greek alphabet")
```
